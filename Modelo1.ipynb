```python
import pyreadr
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
data = pyreadr.read_r("loan_data_ch1.rds")[None]
```

```python
data.head()
```

```python
plt.hist(data[["int_rate"]], color = "white", edgecolor = "black", bins = 19)
plt.xlabel("Interest rate")
plt.ylabel("Frequency")
plt.title("Histogram of interest rate")
plt.show()
```

```python
print(len(data))
print(sqrt(len(data)))
```

```python
n, bins, patches = plt.hist(data[["annual_inc"]], color = "white", edgecolor = "black", bins = int(sqrt(len(data))))
plt.xscale("log")
plt.xlabel("Annual income")
plt.ylabel("Frequency")
plt.title("Histogram of interest rate")
plt.show()
```

```python
print(bins)
```

```python
print("Detalles de cada bin:")
for i in range(len(bins)-1):
    print(f"Bin {i+1}: [{bins[i]:.2f}, {bins[i+1]:.2f}] - frecuencia: {n[i]}")
```
